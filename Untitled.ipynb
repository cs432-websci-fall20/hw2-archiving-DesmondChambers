```python
import json
import os
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
mementos = []
#mementos.append(0)
for file in os.listdir('./timemaps/'):
    try:
        with open(f"./timemaps/{file}", "r") as f:
            data = json.load(f)

            #print(len(data['mementos']['list']))
            mementos.append(len(data['mementos']['list']))
            #print(data['mementos']['list'])
            #print()

    except:
        continue
```

```python
plt.hist(mementos, ec="red")
plt.xlabel("number of mementos")
plt.ylabel("frequency of occurence")
plt.title("TimeMaps")
plt.show()
```

```python
mementos_dates = []
#mementos_dates.append(0)
for file in os.listdir('./timemaps/'):
    try:
        with open(f"./timemaps/{file}", "r") as f:
            data = json.load(f)

            d1 = datetime.strptime(data['mementos']['first']['datetime'].split("T")[0], "%Y-%m-%d")
            d2 = datetime.strptime(data['mementos']['last']['datetime'].split("T")[0], "%Y-%m-%d")
            mementos_dates.append((d2 - d1).days)

    except:
        continue
```

```python
plt.scatter(mementos_dates, mementos)
plt.xlabel("age")
plt.ylabel("number of mementos")
plt.title("Q2")
plt.show()
```
